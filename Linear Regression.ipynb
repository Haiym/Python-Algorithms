```python
import pandas as pd
import random as rnd
import csv
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
# Generate 50 random numbers between 30 and 50
x = [rnd.randint(30, 50) for _ in range(50)]
# Generate 50 random numbers between 1 and 100 for the new column
y = [rnd.randint(1, 30) for _ in range(50)]

# Define the file name
csv_file = "homeprices.csv"

# Write the random numbers to the CSV file with the "age" and "other_column" column headers
with open(csv_file, mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['age', 'hours'])  # Write the column headers
    for age, hour in zip(x, y):
        writer.writerow([age, hour])

print("CSV file created successfully!")
```
