```python
import numpy as np
import os
os.environ['OMP_NUM_THREADS'] = '1'  # Set environment variable to avoid memory leak issue
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
# Generate synthetic data for customer segmentation
np.random.seed(0)
# Generate 200 data points with two features: spending score (x) and annual income (y)
X = np.random.rand(200, 2) * 100
X
```

```python
# Visualize the synthetic data
plt.scatter(X[:, 0], X[:, 1])
plt.xlabel('Spending Score')
plt.ylabel('Annual Income')
plt.title('Customer Segmentation Data')
plt.show()
```

```python
# Perform K-Means clustering with 5 clusters
kms= KMeans(n_clusters=5, n_init=10)
kms.fit(X)
```

```python
# Get the cluster centroids and labels
centroids=kms.cluster_centers_
labels=kms.labels_
```

```python
# Visualize the clusters along with their centroids
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', alpha=0.5)
plt.scatter(centroids[:, 0], centroids[:, 1], marker='x', c='red', s=200, label='Centroids')
plt.xlabel('Spending Score')
plt.ylabel('Annual Income')
plt.title('Customer Segmentation with K-Means Clustering')
plt.legend()
plt.show()

# Print the cluster centers and labels
print("Cluster Centers:")
print(centroids)
print("\nLabels:")
print(labels)
```
